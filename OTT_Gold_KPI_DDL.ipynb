```python
%sql
CREATE TABLE IF NOT EXISTS ott_catalog.gold.kpi_content_snapshot(
    snapshot_month STRING,
    total_titles BIGINT,
    total_movies BIGINT,
    total_series BIGINT,
    avg_movie_duration DECIMAL(18,2),
    avg_series_seasons DECIMAL(18,2),
    pct_movies DECIMAL(18,2),
    pct_series DECIMAL(18,2)

)
USING DELTA
```

```python
%sql
CREATE TABLE IF NOT EXISTS ott_catalog.gold.kpi_content_rating (
    snapshot_month STRING,
    rating STRING,
    titles_count BIGINT,
    movies_count BIGINT,
    series_count BIGINT,
    pct_of_total_titles DECIMAL(18,2),
    pct_new_titles DECIMAL(18,2)
)
USING DELTA;
```
