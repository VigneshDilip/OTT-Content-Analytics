### Transformation_Bronze_To_Silver

```python
dbutils.widgets.text('table_name','ott_directors')
```

```python
tablename=dbutils.widgets.get('table_name')
```

```python
ott_df=spark.read.format('delta').load(f'abfss://bronze@storageottanalytics.dfs.core.windows.net/{tablename}')
ott_df=ott_df.drop(ott_df['_rescued_data']).where(ott_df['show_id'] !='None')
ott_df.display()
```

```python
ott_df.write.format('delta').mode('append').option('mergeSchema','true').option('path',f'abfss://silver@storageottanalytics.dfs.core.windows.net/{tablename}').save()
```
